#Stability of CuSbS2


```python
import numpy as np
%pylab inline
p0 = 1e5
R2 = 8.314 #J/mol K
R = 8.6e-5 #ev/K
```

# 1. Enthalpy of Decompostion

```python
dH_d = 3.09 #eV
dH0 = 21
dH_d = dH0 *36*4/1000. #eV
k_mmhg_pa = 133.32

T= np.linspace(500,1000,21)

Psb2s3 = 10**(13.96 - (10490./T))


Psb4 = 10**( 11.539 - (11025/T)) * k_mmhg_pa 
Psb4 = 10**( 10.545  - (9913./T)) 


dG1 = -R * T * np.log(Psb2s3/p0)
dG2 = -R * T * np.log(Psb4/p0)
print dG1, dG2, Psb2s3, T

```

```python
dG = dH_d + 0.5 * dG2 + 11 * dG1
```

```python
11/(36*4.)
```

```python

K = np.e**(-dG/(T*R))
Psb4_sat = (K/22.**11)**(1/11.5) *p0
plot(T-273, 22*Psb4_sat, '-b')
plot(T-273, Psb4_sat, 'g')
plot(T-273, Psb2s3, 'k')
ax = gca()

plot([100,800],[0.4,0.4], '--k')
ax.set_title(r'Correct, $\Delta H = %d $ meV'%dH0 )
ax.axis(xmin=300, xmax=700, ymin=1e-10, ymax=1.0)
ax.set_xlabel('$T$ [$^{\circ}\mathrm{C}$]')
ax.set_ylabel('$p$ [Pa]')
ax.set_yscale('log',basey=10)
#legend()
```
