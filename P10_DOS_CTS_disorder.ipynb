```python
def calc_ipr(pdos):
    Natoms, Nel = np.shape(pdos)
    c2 = pdos * pdos
    c4 = c2 * c2
    ipr = Natoms * np.sum(c4, axis=0)/np.sum(c2, axis=0)**2
    return ipr
#ipr = np.nan_to_num(ipr)
```

```python

def get_pdos(DOS, N, Nat):
    pdos = np.zeros([Nat,N])
    for i in range(Nat):
        for id in range(9):
            pdos[i] += DOS.site_dos(i, id)
    return pdos
```

```python
def get_dos_czts(ncu3sn=0):
    N=2501
    Nat = 216
    cutoff_min, cutoff_max = -20.0, -16.0
    
    path = '/scratch/pzawadzk/DISORDER/CZTS/cts_mix_3/cu_%d_%d/dos/DOSCAR'
    path_ref = '/scratch/pzawadzk/DISORDER/CZTS/cts_mix_3/cu_%d_%d/non-magnetic/DOSCAR'
    dos = np.zeros(N)
    count = 0
    pdos = np.zeros([Nat,N])
    seeds = [1,2, 3, 4]
    pdos_ref = zeros(301)
    avg = 0
    weight = 0
    cbms = []
    vbms = []
    for seed in seeds:
        if 1:#try:
            avg = 0
            weight = 0
            calc = Extract(path[:-7]%(ncu3sn, seed))
            vbms.append(calc.vbm)
            cbms.append(calc.cbm)
            mult = float(1e3/calc.volume)
            DOS = VaspDos(path%(ncu3sn, seed))
            dos += DOS.dos*mult
            count += 1
            pdos += get_pdos(DOS, N, Nat)
            
            DOS_ref = VaspDos(path_ref%(ncu3sn, seed))
            for i, atom in enumerate(calc.structure):
                  if atom.type == 'Sn':
                    for id in [4,5,6,7,8]:#d
                      pdos_ref += DOS_ref.site_dos(i, id)
            for i in range(301):
                if DOS_ref.energy[i] > cutoff_min and DOS_ref.energy[i] < cutoff_max:
                    avg += DOS_ref.energy[i] * pdos_ref[i]
                    weight += pdos_ref[i]
        #except:
            avg /= weight
            print avg
            print 'ERROR', seed
    

    avg /= weight
    
    dos /= count
    pdos /= count
    return dos, pdos, DOS.energy, avg, cbms, vbms
```
