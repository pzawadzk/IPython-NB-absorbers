```python
import numpy as np
%pylab inline
from pylada.vasp import Extract, MassExtract
from pylada.crystal import  neighbors
import os
```

```python
dir0 = '/scratch/pzawadzk/DISORDER/mat_mot/data'
```

```python
anions = ['S', 'N', 'O', 'P']
dirs = ['N_Al_Cu_Sn', 
        'O_Al_Cu_Sn', 'O_Al_Cu_Zn', 'O_Cu_Sn_Zn', 
        'P_Al_Sb_Zn', 'P_Al_Sn_Zn', 'P_Cu_Sb_Zn', 'P_Cu_Sn_Zn', 'P_Sb_Sn_Zn', 
        'S_Al_Cu_Sn', 'S_Al_Cu_Zn', 'S_Cu_Sn_Zn']


q_dict = {'S':6, 'N': 5, 'O':6, 'P':5,
          'Al': 3, 'Cu':1, 'Sb':3, 'Sn':4, 'Zn':2 }
elneg_dict = {'S':2.5, 'N': 3.0, 'O': 3.5, 'P':2.1,
          'Al': 1.5, 'Cu':1.9, 'Sb':1.9, 'Sn':1.8, 'Zn':1.6 }
```

```python
def get_coor_shell(strucutre, atom):
    neighs = [n for n in neighbors(strucutre, 4, atom.pos)]
    atom_list = []
   # d_tmp = 10
    for n in neighs:
       # print n
       # d = float(n[-1]*strucutre.scale)
      #  if (d-d_tmp) > 0.5:
      #      break
      #  else:
            atom_list.append(n[0].type)
       #     d_tmp = d
    return atom_list
        
```

```python
def get_el_count(structure):
    el_count_tot = 0
    for atom in structure:
        if atom.type in anions:
            coords =  get_coor_shell(structure, atom)
            el_count = q_dict[atom.type] - 8
            avg_eln = 0
            for symbol in coords:
                el_count += q_dict[symbol]/4.
                avg_eln += elneg_dict[symbol]/4.
            el_count_tot += abs(el_count)*(elneg_dict[atom.type] - avg_eln)
    return el_count_tot
            
    
            
```

```python
for dir_i in dirs:
    dir1 = os.path.join(dir0, dir_i)
    for dir_j in os.listdir(dir1):
        calcs = MassExtract(os.path.join(dir1, dir_j))
        absNs = []
        Es = []
        for key in calcs.keys():
            if 'relax' not in key:
    
                absN = get_el_count(calcs[key].structure)
                absNs.append(absN)
                E = calcs[key].energy_sigma0
                Es.append(E)
        figure()
        ax = gca()
        ax.set_title('%s__%s' %(dir_i, dir_j))
        ax.plot(Es, absNs, 'x')
```
