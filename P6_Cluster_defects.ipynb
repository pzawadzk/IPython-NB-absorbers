# Calculates cluster of motifs around a defect

```python
import os
os.chdir('P6')
```

```python
from pylada.crystal import read, write, neighbors, supercell
import numpy as np
from IPython.display import Image
%pylab inline
local = True
```

```python
if not local:
    structure = read.poscar('/scratch/pzawadzk/materials/Cu2SnS3/non-magnetic/POSCAR')
    write.poscar(structure, 'POSCAR_CTS.vasp', vasp5=True)
else:
    structure = read.poscar('POSCAR_CTS.vasp')

k = [2,2,2]
cell = structure.cell
scell = np.array([[k[0], 0, 0], [0, k[1], 0], [0, 0, k[2]]])
cell = np.dot(cell, scell)
structure  = supercell(structure, cell)
print len(structure)/2.
```

```python
sn_atoms = []
for atom in structure:
    if atom.type == 'Sn':
        sn_atoms.append(atom)
```

```python
for i, atom in enumerate(sn_atoms):
    print i, atom.pos
```

```python
index = 6
sn_atoms[index].type = 'Cu'
neighs = [n for n in neighbors(structure, 4, sn_atoms[index].pos)]
atoms_to_keep = [sn_atoms[index]]
for n in neighs:
    atoms_to_keep.append(n[0])
    neighs_2nd = [n for n in neighbors(structure, 4, n[0].pos)]
    for n_2nd in neighs_2nd:
        if n_2nd[0] not in atoms_to_keep:
            atoms_to_keep.append(n_2nd[0])
```

```python
len(atoms_to_keep)
```

```python
str_tmp = structure.copy()
N = len(structure)
for i, atom in enumerate(structure[::-1]):
    if atom not in atoms_to_keep:
        str_tmp.pop(N-1-i)
write.poscar(str_tmp, '/home/pzawadzk/cu_sn_cts.vasp', vasp5=True)
```
