```python
import numpy as np
import os
```

```python
N_sites = 48*8
N_points = N_sites
zn_dat = np.linspace(0,N_sites,N_points+1)
```

```python
points = [[2*N_sites/3, N_sites/3, 0], [3*N_sites/4, N_sites/4, 0], [0,0, N_sites], [N_sites/2, N_sites/4, N_sites/4]]
xs = [[100,0,0, 1],[0,100,0, 1],[0,0,100, 1], [50, 0, 50, 1]]


d_cut = 0.109

for zn in zn_dat:
   # print '###', zn
    min_sn = (N_sites-zn)/4.
    max_sn = (N_sites-zn)/3.
    for sn in np.linspace(min_sn, max_sn, (max_sn-min_sn)//1 +1):
        cu = N_sites - zn -  sn
        point =  [cu, sn, zn]
        n = zn + sn
        x = [100*(3*sn - cu)/n, 100*(cu - 2 * sn)/n, 100*zn/n, np.random.random_sample()*2-1 ] #CTS, 314 ,ZnS

        check = True
        for x_tmp in xs:
            dx = abs(np.array(x) - np.array(x_tmp))/100.
            d = abs(np.dot(dx, dx))**0.5
            if d < d_cut:
                check = False
                break
        if check and sn - int(sn) == 0:
           # print sn, cu, zn      
            xs.append(x)
            points.append(point)
print len(points)
```

### Mixing from the model H

```python
E0_ZnS = -208.586/96.
E0_CTS = -94.857/32.
E0_314 = -98.212/24.
N = 384
dir = '/scratch/pzawadzk/DISORDER/CZTS/MC/CZTS/alloy'
all_points = []
for path in os.listdir(dir):
    if '%d_T_'%N in path:
        #print path
        split_path = path.split('_')
        #print split_path
        file = os.path.join(dir, path, 'mc_N_Cu_Zn_Sn_.dat')
        data = np.loadtxt(file)
        n_cu = float(split_path[3])
        n_sn = float(split_path[4])
        n_zn = float(split_path[5])
        n = n_zn + n_sn
        x_cts = (3*n_sn - n_cu)/n
        x_314 = (n_cu - 2 * n_sn)/n
        x_zns = n_zn/n
        E = data[-1][-5]
        E_mix = E - (x_cts * E0_CTS + x_314 * E0_314 + x_zns * E0_ZnS)*n
        E_mix *= 1000./(2*N)
        #print E_mix, n_cu, n_sn, n_zn
        all_points.append([100*x_cts, 100*x_314, 100*x_zns, E_mix])

all_points = np.array(all_points)
np.savetxt('/home/pzawadzk/alloy.dat', all_points)
```

```python
E0_CZTS = -127.361/24.
E0_CTS = -94.857/32.
E0_314 = -98.212/24.
N = 384
dir = '/scratch/pzawadzk/DISORDER/CZTS/MC/CZTS/alloy'
all_points = []
for path in os.listdir(dir):
    if '%d_T_'%N in path:
        split_path = path.split('_')
        file = os.path.join(dir, path, 'mc_N_Cu_Zn_Sn_.dat')
        data = np.loadtxt(file)
        n_cu = float(split_path[3])
        n_sn = float(split_path[4])
        n_zn = float(split_path[5])
        n =  n_sn
        if n>0:
            x_cts = (3*n_sn - n_cu - n_zn)/n
            x_314 = (n_cu - 2 * n_sn)/n
            x_czts = n_zn/n
            E = data[-1][-5]
            E_mix = E - (x_cts * E0_CTS + x_314 * E0_314 + x_czts * E0_CZTS)*n
            E_mix *= 1000./(2*N)
            if x_cts>=0 and x_314>=0 and x_czts>=0:
                all_points.append([100*x_cts, 100*x_314, 100*x_czts, E_mix])
                #print x_cts, x_314, x_czts

all_points = np.array(all_points)
np.savetxt('data_p100_6/alloy_czts.dat', all_points)
```
