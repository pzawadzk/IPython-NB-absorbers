```python
%pylab inline
import numpy as np
import os
import sys
from tools import count_configurations_5
from pylada.crystal import neighbors, write
from pylada.vasp import Extract, MassExtract
from ase.calculators.vasp import VaspDos
import pickle
```

```python
def get_pdos(dir0, key, calc):
        x, y, Es, Dos = [],[],[],[]
        Pdos_s, Pdos_p = [], []

        vol = calc.volume.__float__()
        dos_path = os.path.join(dir0, key[1:], 'DOSCAR')
        
        dos = VaspDos(dos_path)
        Dos.append(1e3*dos.dos/vol)
        energy = dos.energy 
        Ngrid = len(energy)
        Pdos_s = zeros([3,Ngrid])
        Pdos_p = zeros([3,Ngrid])
        Pdos_ref = zeros([3,Ngrid])

        count, conf_count = count_configurations_5(calc.structure,'S', atoms = ['Cu','Zn','Sn'])
        n_cu_types = np.array([0,0,0])           
        for i, atom in enumerate(calc.structure):
            if atom.type == 'S':
                n_cu = -2
                for n in [n for n in neighbors(calc.structure, 4, atom.pos)]:
                    if n[0].type == 'Cu':
                        n_cu += 1
                n_cu_types[n_cu]+=1
                for id in [0]:
                    Pdos_s[n_cu] += dos.site_dos(i, id) *1e3/vol
                for id in [1,2,3]:
                    Pdos_p[n_cu] += dos.site_dos(i, id) *1e3/vol
        avg=0
        for i in range(3):
            if n_cu_types[i]>0:
                Pdos_s[i] /= n_cu_types[i] 
                Pdos_p[i] /= n_cu_types[i] 
        return energy-avg, Dos[0], Pdos_s, Pdos_p,  float(calc.vbm)
```

```python
dir_dis = '/scratch/pzawadzk/DISORDER/CZTS/MC/CTS/energy_dis_069'
calcs_dis = MassExtract(path=dir_dis, )

dir_cc = '/scratch/pzawadzk/DISORDER/CZTS/MC/CTS/energy_cc_069'
calcs_cc = MassExtract(path=dir_cc, )
```

```python
xcu = '48'
#xcu = '50'

if xcu == '48':
    ratio = 0.66
else:
    ratio = 0.69
    
```

```python
Ngrid = 1601
Pdos_s = zeros([3,Ngrid])
Pdos_p = zeros([3,Ngrid])
keys = calcs_dis.keys()
count = 1
for key in keys:
    if 'dos' in key and xcu in key:
        E, dos, ps, pp, vbm = get_pdos(dir_dis, key, calcs_dis[key])
        Pdos_s += ps
        Pdos_p += pp
        count +=1
Pdos_s /= count
Pdos_p /= count
```
