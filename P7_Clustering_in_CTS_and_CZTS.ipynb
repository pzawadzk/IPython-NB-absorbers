```python
#%%px
def get_coord_distrib(poscar, confs, pdf_flag = False, cn_flag = False, vol0=1):
    structure = read.poscar(poscar)
    vol = np.linalg.det(structure.cell)*structure.scale**3/len(structure)
    if vol0 == 1:
        vol0 = vol
    scale = (vol0/vol)**(1/3.)
    structure.scale *= scale
    vol = np.linalg.det(structure.cell)*structure.scale**3/len(structure)
    print vol
    
    structure = remove_atoms(structure, confs = confs)
    hist = np.zeros(13)
    for k, atom in enumerate(structure):
        neighs= [n for n in neighbors(structure, 12, structure[k].pos)]
        count = 0
        for n in neighs:
            if n[-1]*structure.scale <4.9:
                count += 1
        hist[count] += 1
    
    if pdf_flag:
        pdf, r = calc_pdfs(structure, pdf_width=.2, pdf_sigma=0.4, pdf_cut=23, smear_flag=True)
    else:
        pdf, r = {'SS':0}, 0
        
    if cn_flag:   
        cn = int_cn_from_pdfs(structure, pdf, pdf_width=.2)
    else:
        cn = {'SS':0}
    return hist, pdf, r, cn
    
```

```python

def remove_atoms(structure, conf = [3,1,0],  remove_cluster=True, remove_metal=True):
    structure_tmp = structure.copy()
    
    if remove_cluster:  
        atoms_notto_remove = []
        
        index_list = range(len(structure))
        for id in index_list:
            atom = structure[id]
            if atom.type == 'S':
                atoms_tmp = [atom]
                p = atom.pos
                neighs = [n for n in neighbors(structure, 4, p)]
                count_Cu = 0
                count_Sn = 0
                count_Zn = 0
                for n in neighs:
                    atoms_tmp.append(n[0])
                    if n[0].type == 'Cu':
                        count_Cu  +=1
                    elif n[0].type == 'Sn':
                        count_Sn  +=1
                    elif n[0].type == 'Zn':
                        count_Zn  +=1
                    
                if count_Cu == conf[0] and count_Sn == conf[1] and count_Zn == conf[2]:
                    atoms_notto_remove += atoms_tmp
                else:
                    pass
          
        N = len(structure)
        for i in range(N)[::-1]:
            atom = structure[i]
            if atom in atoms_notto_remove:
                pass
            else:
                structure_tmp.pop(i)
    if remove_metal:
        N = len(structure_tmp)
        for i in range(N)[::-1]:
            if structure_tmp[i].type != 'S':
                structure_tmp.pop(i)
    return structure_tmp
                
```
