```python


def get_data_2(dir, frac=0.5):
    data_avg = []
    data_all ={}
    T = []
    path = os.path.join(dir,'mc_N_Cu_Zn_Sn_.dat')
    data = np.loadtxt(path)
    for el in data:
        if el[-1] not in T:
            T.append(el[-1])
            data_all[el[-1]]=[]
        data_all[el[-1]].append(el[1:16])
        
    for t in T:
        slice_dat = data_all[t]
        N = len(slice_dat)
        data_avg.append(np.mean(slice_dat[N//2:], axis=0))
    data_avg = np.array(data_avg)/np.sum(data_avg[0])
    return T, data_avg
    

```

```python

def func(p, x):
  
  f  = np.exp(-((x-p[1])/p[0])**2)/(np.pi**2 *p[0])
  return f

def residuals(p, y, x):
  err = y - func(p, x)
  return err
```
