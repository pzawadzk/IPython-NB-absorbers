```python
#monoclincic
structure = read.poscar('/scratch/pzawadzk/materials/Cu2SnS3/non-magnetic/POSCAR')
k = [3,3,3]
cell = structure.cell
scell = np.array([[k[0], 0, 0], [0, k[1], 0], [0, 0, k[2]]])
cell = np.dot(cell, scell)
structure  = supercell(structure, cell)
vol = np.linalg.det(structure.cell)*structure.scale**3/len(structure)
print vol
N = len(structure)
for i in range(N)[::-1]:
    if structure[i].type == 'S':
        structure.pop(i)
pdf, r = calc_pdfs(structure, pdf_width=.2, pdf_sigma=0.4, pdf_cut=12, smear_flag=True)
cn = int_cn_from_pdfs(structure, pdf, pdf_width=.2)


#disordered
structure_d = read.poscar('/scratch/pzawadzk/DISORDER/CZTS/MC/CTS/data/cu128_192_T_-200.0_seed_4_mc_v5/poscar_N_Cu_Zn_Sn_.vasp')
vol_d = np.linalg.det(structure_d.cell)*structure_d.scale**3/len(structure_d)

x_scale = (vol/vol_d)**(0.3333)
structure_d.scale *= x_scale

print x_scale
N = len(structure_d)

for i in range(N)[::-1]:
    if structure_d[i].type == 'S':
        structure_d.pop(i)
pdf_d, r_d = calc_pdfs(structure_d, pdf_width=.2, pdf_sigma=0.4, pdf_cut=12, smear_flag=True)
cn_d = int_cn_from_pdfs(structure_d, pdf_d, pdf_width=.2)
```
