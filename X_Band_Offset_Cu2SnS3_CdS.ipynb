```python
def movingaverage(interval, N):
  #  window = np.kaiser(int(N), 8)/float(window_size)
  #  avg = np.convolve(interval, window, 'same')
    avg = np.convolve(interval, np.ones((N,))/N)[(N-1):]
    return avg
```

```python
def get_pot(path):

    from pylada.vasp import Extract
    from matplotlib import rc
    import matplotlib
    import matplotlib.pyplot as plt
    from ase.calculators.vasp import VaspChargeDensity
    from ase.io import read

    vasp_charge = VaspChargeDensity(path+'/LOCPOT')
    chg = vasp_charge.chg[0]

    Nx, Ny, Nz = np.shape(chg)
    chg = vasp_charge.chg[0]
    vol = vasp_charge.atoms[0].get_volume()
    vz  = vol*chg.sum(0).sum(0)/(Nx*Ny)
    vz = np.concatenate([vz, vz])

    zmax = vasp_charge.atoms[0].cell[2,2] * 2
    Nz *= 2
    z = np.linspace(0, zmax, Nz)

    return z, vz, zmax
```

```python

def plot_fit(v, z, zmax, start, end):
    
    
    istart = int(len(z)*start/zmax)
    iend = int(len(z)*end/zmax)
    p1 = np.polyfit(z[istart:iend], v[istart:iend], 1) 

    t=np.linspace(start,end,10)
    y1 = p1[0]*t + p1[1]
   
    plot(t, y1, '-k')
    print p1
    return p1
   
```

```python
data_file = 'data/band_offset_cts.pckl'
if os.path.exists(data_file):
    f = open(data_file, 'r')
    data = pickle.load(f)
else:
    data = {}
    path = '/scratch/pzawadzk/surfaces/CTS/epi_pere'    
    extract = MassExtract(path, naked_end=True)
    
    for data_path in extract.keys():
        try:
            line = data_path.split('/')   
            z, vz, zmax = get_pot(os.path.join(path, line[1], line[2]))
            data[data_path] = [z, vz, zmax]
        except:
            print 'error in', data_path
    f = open(data_file, 'w')
    pickle.dump(data, f)

f.close()
```

```python
def dipole_moment(slab=None,charge=None):
    """Claculates the dipole moment"""
    z=[]
    ch=[]
    for atom in slab:
        z.append(atom.pos[2])
        ch.append(charge[atom.type])
    assert not abs(sum(array(ch)))>1e-5, "System is not charge neutral!"
    dipole_moment=sum(array(ch)*array(z))
    return dipole_moment, z, ch
```
