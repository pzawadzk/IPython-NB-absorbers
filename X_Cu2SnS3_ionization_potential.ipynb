```python
def movingaverage(interval, window_size):
    window = np.ones(int(window_size))/float(window_size)
    return np.convolve(interval, window, 'same')
```

```python

def get_pot_alligment(dir0, n_lay, mult, window, delta):
    dir = os.path.join(dir0, key, 'Cu2SnS3', 'lay_%d/non-magnetic'% n_lay)
    locpot = os.path.join(dir, 'LOCPOT')
    #locpot = os.path.join(dir, 'scaled', 'LOCPOT')
    poscar  = os.path.join(dir, 'OUTCAR')
    #poscar  = os.path.join(dir, 'scaled', 'OUTCAR')
    
    str = read(poscar)
    
    top_nr = max(str.positions[:,2])
    bottom_nr = min(str.positions[:,2])
    z_layer = (top_nr-bottom_nr)/(n_lay*mult)
    #print z_layer, top_nr, bottom_nr
    
    vasp_charge = VaspChargeDensity(locpot)
    chg = vasp_charge.chg[0]
    Nx, Ny, Nz = np.shape(chg)
    
    chg = vasp_charge.chg[0]
    
    vol = vasp_charge.atoms[0].get_volume()
    vz  = vol*chg.sum(0).sum(0)/(Nx*Ny)
    vz  -= vz[0]
    
    zmax = vasp_charge.atoms[0].cell[2,2]
    
    top = max(vasp_charge.atoms[0].positions[:,2])
    bottom  = min(vasp_charge.atoms[0].positions[:,2])
    
    z = np.linspace(0, zmax, Nz)
    
    
    
    w = int(window * Nz * z_layer/zmax)
    d = int(delta * Nz * z_layer/zmax)
    t = int(top * Nz /zmax) 
    b = int(bottom * Nz /zmax) 
    
    top = t*zmax/Nz
    bottom = b*zmax/Nz 
    delta = d*zmax/Nz
    
    vz_avg = movingaverage(vz, w)
    vz_avg_range = vz_avg[b+d:t-d]
    ref = sum(vz_avg_range)/(len(vz_avg_range))
    
    return ref, z, vz, vz_avg, top, bottom, delta
```
