```python
from ase.calculators.vasp import VaspDos
from pylada.vasp import Extract

data_file = 'data/motifs_dos.pckl'
if 0:#os.path.exists(data_file):
    f = open(data_file, 'r')
    Efs, Evbm, Ecbm, Pdos,  Es, Dos, Avgs = pickle.load(f)
else:  
    dirs = ['cu3sns4','cu2znsns4','cuzn2sns4']	

    cutoff_min, cutoff_max = -20.0, -16.0
    cutoff_min, cutoff_max = -12.0, -7.0

    Es = []
    Dos = []
    Pdos = []
    Pdos_Zn_s = []
    Efs = []
    Evbm = []
    Ecbm = []
    Avgs = []
    for k, dir in enumerate(dirs):
        calc = Extract(os.path.join('/scratch/pzawadzk/DISORDER/CZTS/phase/dos', dir, 'dos','OUTCAR'))
        vol = calc.volume.__float__()
        dos = VaspDos(os.path.join('/scratch/pzawadzk/DISORDER/CZTS/phase/dos', dir, 'dos','DOSCAR'))
        Dos.append(1e3*dos.dos/vol)
        energy = dos.energy 
        Ngrid = len(energy)
        pdos_s = zeros(Ngrid)
        pdos_Zn_s = zeros(Ngrid)
        for i, atom in enumerate(calc.structure):
            #if atom.type == 'Sn':
             #   for id in [4,5,6,7,8]:
              #      pdos_s += dos.site_dos(i, id)
            if atom.type == 'S':
                for id in [0]:
                    pdos_s += dos.site_dos(i, id)
        if atom.type == 'Sn':
			for id in [0]:
				pdos_Zn_s += dos.site_dos(i, id)
                    
        avg = 0
        weight = 0
        Pdos.append(1e3*pdos_s/vol)
        Pdos_Zn_s.append(1e3*pdos_Zn_s/vol)
        for i in range(Ngrid):
            if energy[i] < cutoff_max and energy[i] > cutoff_min:
                avg += energy[i] * pdos_s[i]
                weight +=	pdos_s[i]
        avg /= weight
        print dir, avg
        avg =  Avgs_ref[k]
        Ef =	float(calc.fermi_energy) - avg #- calc.vbm.__float__()
        Efs.append(Ef)
        print dir, avg, Ef
        #Avgs.append(avg)
        Es.append(energy-avg)
        Evbm.append( calc.vbm.__float__()-avg)
        Ecbm.append( calc.cbm.__float__()-avg)
    f = open(data_file, 'w')
    pickle.dump([Efs, Evbm, Ecbm, Pdos,  Es, Dos, Avgs], f)
f.close()
```
